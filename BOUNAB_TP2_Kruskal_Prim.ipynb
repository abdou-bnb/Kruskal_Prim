```python
#TP2_BOUNAB_Abdelmounaim

#Principe de l'ALGO DE KRUSKAL : d'abord je vais trier les arrets par poids croissants et a chaque fois je vais voir l'arret courant et l'ajouter a l'arbre si elle ne pose pas un cycle.

#Principe de l'ALGO DE PRIM  : d'abord je vais choisir n'importe quelle sommet pour commencer, a chaque fois je vais ajouter un sommet et une arrete avec une poids minimale et eviter les cycles.

class Graph:
    def __init__(self, vertices):
        self.V = vertices
        self.graph = []  # Liste pour stocker les arêtes du graphe



    def prim_algorithm(self):
        parent = [-1] * self.V  # Tableau pour stocker les sommets parents de l'arbre couvrant minimal
        key = [float('inf')] * self.V  # Tableau pour stocker les clés (poids minimaux) des sommets
        key[0] = 0  # La clé du premier sommet (0) est initialisée à 0
        mst_set = [False] * self.V  # Liste pour suivre les sommets inclus dans l'arbre couvrant

        for _ in range(self.V):
            min_key = float('inf')
            for v in range(self.V):
                # Trouver le sommet non inclus avec la clé minimale
                if not mst_set[v] and key[v] < min_key:
                    min_key = key[v]
                    min_index = v

            mst_set[min_index] = True  # Marquer le sommet comme inclus

            for u, v, w in self.graph:
                # Mettre à jour les clés des sommets adjacents non inclus si nécessaire
                if not mst_set[v] and u == min_index and w < key[v]:
                    parent[v] = u
                    key[v] = w

        print("Arbre couvrant de poids minimal (Prim) :")
        for i in range(1, self.V):
            print(f"{chr(65 + parent[i])} - {chr(65 + i)} : {key[i]}")


    def add_edge(self, u, v, w):

        #Méthode pour ajouter une arête au graphe avec ses sommets (u et v) et son poids (w).
        self.graph.append([u, v, w])


    def find(self, parent, i):
        # Fonction pour trouver le représentant (parent) d'un ensemble
        if parent[i] == i:
            return i
        return self.find(parent, parent[i])


    def kruskal_algorithm(self):
        self.graph.sort(key=lambda item: item[2])  # Trier les arêtes par poids croissant
        parent = [i for i in range(self.V)]  # Initialiser chaque sommet comme son propre parent
        result = []
        i = 0




        while len(result) < self.V - 1:
            u, v, w = self.graph[i]
            i += 1
            x = self.find(parent, u)
            y = self.find(parent, v)
            if x != y:
                result.append([u, v, w])
                self.union(parent, x, y)

        print("Arbre couvrant de poids minimal (Kruskal) :")
        for u, v, weight in result:
            print(f"{chr(65 + u)} - {chr(65 + v)} : {weight}")



    def union(self, parent, x, y):
        # Fonction pour fusionner deux ensembles en mettant à jour le parent
        x_set = self.find(parent, x)
        y_set = self.find(parent, y)
        parent[x_set] = y_set


# Création d'un nouveau graphe
g = Graph(7)

# Importation des données depuis le fichier texte
with open('graph_TP2.txt', 'r') as file:
    lines = file.readlines()
    for line in lines:
        u, v, w = line.strip().split()
        u = ord(u) - ord('A')
        v = ord(v) - ord('A')
        w = int(w)
        g.add_edge(u, v, w)

for edge in edges:
    u, v, w = edge
    graphe7.add_edge(ord(u) - ord('A'), ord(v) - ord('A'), w)

graphe7.prim_algorithm()
graphe7.kruskal_algorithm()
```
